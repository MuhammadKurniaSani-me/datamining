# 📚 Penambangan Data


## Apa itu penambangan data 🔎❓


* ➡️  Metode mengambil, membersihkan, memproses data untuk diolah menjadi sebuah informasi yang bermanfaat
* ➡️  Mata kuliah yang mempelajari proses pengelolahan data mentah menjadi data jadi dengan algoritma tertentu
* ➡️  Penambangan data dapat digunakan untuk melakukan klasifikasi data dengan algoritma yang tepat
